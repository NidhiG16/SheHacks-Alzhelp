```python
import pandas as pd
import numpy as np
import datetime
from time import strftime
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as  sns
```

```python
data = pd.read_csv('sample data1.csv')
```

```python
data.info()
```

```python
data['Date'] = pd.to_datetime(data['Date']).dt.date.astype('datetime64[ns]')
```

```python
data.info()
```

```python
data['Show/No show'].value_counts().plot(kind = 'barh', figsize=(8,6))
plt.xlabel("daily checkup", labelpad=14)
plt.ylabel("attended", labelpad=14)
plt.title("frequency of checkups attended by the patient", y=1.02);
```

```python
plt.figure(figsize=(20,10))
plt.xticks(rotation=90)
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df.Date, format='%Y-%m-%d')
d = df[df['Date'].dt.strftime('%m')=='05']
sns.barplot(d['Date'], d['weight'])
plt.show()
```

```python
sns.distplot(data['MMSE'])
```

```python
sns.jointplot(data['Date'], data['weight'])
```

```python
sns.jointplot(data['diastolic BP'], data['systolic BP'])
```

```python
sns.pairplot(data[['weight', 'systolic BP']])
```

```python
plt.figure(figsize=(10,10))
plt.yticks(rotation = 90)
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df.Date, format='%d-%m-%Y')
d = df[df['Date'].dt.strftime('%m-%Y')=='05-2023']

sns.stripplot(d['MMSE'], d['Date'])
plt.show()
```

```python
df = pd.read_csv('sample data1.csv', parse_dates=['Date'], index_col='Date')
df_3m = df.groupby(pd.Grouper(freq='3M')).mean()
sns.lineplot(data=df_3m, x=df_3m.index, y=data['weight'])
```

```python
df_grouped = df.resample('3M').sum()

plt.plot(df_grouped.index, df_grouped['weight'])
plt.xticks(rotation = 90)
plt.show()
```

```python
df_grouped = df.resample('3M').sum()

plt.plot(df_grouped.index, df_grouped['MMSE'])
plt.xticks(rotation = 90)
plt.show()
```
