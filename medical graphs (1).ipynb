```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("sample data1 (1).csv")
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df.Date, format='%d-%m-%Y')
d = df[df['Date'].dt.strftime('%m-%Y')=='05-2023']
plt.plot(d['Date'], d['weight'],label='Weight')
plt.plot(d['Date'], d['diastolic BP'],label='diastolic BP')
plt.plot(d['Date'], d['systolic BP'],label='systolic BP')
plt.plot(d['Date'], d['Cholestrol'],label='Cholestrol')
plt.xlabel("Date") 
plt.ylabel("Test result") 
plt.title("variation of test results")
plt.xticks(rotation = 90)
plt.legend()
```

```python
plt.bar(d['Date'], d['MMSE'],label='MMSE')
plt.bar(d['Date'], d['Clinical Dementia Rate'],label='Clinical Dementia Rate')
plt.xlabel("Date") 
plt.ylabel("Test result") 
plt.title("variation of test results")
plt.xticks(rotation = 90)
plt.legend()
```
